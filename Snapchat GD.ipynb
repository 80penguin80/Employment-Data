```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
path = '/Users/zack/Desktop/snap.xls'
df = pd.read_excel(path)
df.columns
```

```python
date=df['As Of Date']
i=0
date_n=[]

for x in date:
    date_n.append(date[i][0:8])
    i=i+1
```

```python
plt.figure(figsize=(25,10))
SM = df['Rating: Senior Management'].expanding().mean()
plt.plot(date_n,SM)

OR= df['Rating: Overall'].expanding().mean()
plt.plot(date_n, OR)

CV= df['Rating: Culture & Values'].expanding().mean()
plt.plot(date_n, CV)

CO= df['Rating: Career Opportunities'].expanding().mean()
plt.plot(date_n, CO)

CB=df['Rating: Comp & Benefits'].expanding().mean()
plt.plot(date_n, CB)
```

```python
plt.figure(figsize=(25,10))
cum_avg=df['Length of Employment'].expanding().mean()
plt.plot(date_n,cum_avg)
plt.xlabel('Date')
plt.ylabel('Years')
plt.title('Avg Length of Employment')
plt.show()
```
